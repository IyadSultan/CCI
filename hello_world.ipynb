## 2. Variables & Data Types 📦

A **variable** is a named container that stores a value. Python figures out the type automatically.

| Type | Example | Description |
|------|---------|-------------|
| `int` | `42` | Whole numbers |
| `float` | `3.14` | Decimal numbers |
| `str` | `"hello"` | Text |
| `bool` | `True` / `False` | Logical values |

````python
# Assigning variables
name = "Alice"          # str
age = 30                # int
height = 1.68           # float
is_student = True       # bool

# Printing with type information
print(f"Name:       {name}  → type: {type(name).__name__}")
print(f"Age:        {age}   → type: {type(age).__name__}")
print(f"Height:     {height}  → type: {type(height).__name__}")
print(f"Is student: {is_student} → type: {type(is_student).__name__}")
````

> 🔍 **Note:** `type()` returns the data type of any variable. `f"..."` is an **f-string** — a convenient way to embed variables inside text.

## 3. Operators ➕➖✖️➗

Python supports arithmetic, comparison, and logical operators.

### Arithmetic Operators
| Operator | Meaning | Example |
|----------|---------|--------|
| `+` | Addition | `5 + 3 → 8` |
| `-` | Subtraction | `5 - 3 → 2` |
| `*` | Multiplication | `5 * 3 → 15` |
| `/` | Division | `5 / 2 → 2.5` |
| `//` | Floor Division | `5 // 2 → 2` |
| `%` | Modulus (remainder) | `5 % 2 → 1` |
| `**` | Exponent | `2 ** 3 → 8` |

````python
a = 10
b = 3

print(f"{a} + {b}  = {a + b}")
print(f"{a} - {b}  = {a - b}")
print(f"{a} * {b}  = {a * b}")
print(f"{a} / {b}  = {a / b:.2f}")
print(f"{a} // {b} = {a // b}  (floor division)")
print(f"{a} % {b}  = {a % b}   (remainder)")
print(f"{a} ** {b} = {a ** b}  (power)")
````

````python
# Comparison operators → return True or False
x = 7
print(f"x == 7 : {x == 7}")
print(f"x != 5 : {x != 5}")
print(f"x > 10 : {x > 10}")
print(f"x <= 7 : {x <= 7}")

# Logical operators
print(f"\nTrue and False : {True and False}")
print(f"True or False  : {True or False}")
print(f"not True       : {not True}")
````

## 4. Strings 🔤

Strings are sequences of characters. Python offers many built-in tools for working with them.

````python
sentence = "Python is powerful and fun!"

print(sentence.upper())          # Convert to uppercase
print(sentence.lower())          # Convert to lowercase
print(sentence.replace("fun", "amazing"))  # Replace a word
print(sentence.split(" "))       # Split into a list of words
print(f"Length: {len(sentence)} characters")
print(f"Starts with 'Python': {sentence.startswith('Python')}")
````

````python
# String indexing and slicing
word = "Python"
#        P  y  t  h  o  n
#        0  1  2  3  4  5   ← positive indices
#       -6 -5 -4 -3 -2 -1   ← negative indices

print(f"First character : {word[0]}")
print(f"Last character  : {word[-1]}")
print(f"Slice [0:3]     : {word[0:3]}")
print(f"Reversed        : {word[::-1]}")
````

## 5. Lists & Tuples 📋

- **List** — ordered, mutable (changeable) collection: `[1, 2, 3]`
- **Tuple** — ordered, immutable (fixed) collection: `(1, 2, 3)`

````python
# Lists
fruits = ["apple", "banana", "cherry"]

print("Original list:", fruits)
fruits.append("mango")           # Add to end
print("After append:", fruits)
fruits.remove("banana")          # Remove by value
print("After remove:", fruits)
print("Item at index 1:", fruits[1])
print("Length:", len(fruits))
print("Sorted:", sorted(fruits))
````

````python
# Tuples — use when data should not change
coordinates = (40.7128, -74.0060)  # Latitude, Longitude of New York
print("Coordinates:", coordinates)
print("Latitude:", coordinates[0])

# Tuples cannot be modified — this would raise an error:
# coordinates[0] = 99  # ← TypeError!
````

## 6. Dictionaries 📖

A **dictionary** stores data as **key-value pairs** — like a real dictionary where each word (key) has a definition (value).

**Syntax:** `{key: value, key: value, ...}`

````python
patient = {
    "name": "Sara",
    "age": 45,
    "diagnosis": "Type 2 Diabetes",
    "active": True
}

# Access values by key
print("Patient name:", patient["name"])
print("Age:", patient["age"])

# Add or update a key
patient["ward"] = "Endocrinology"
print("\nUpdated record:")
for key, value in patient.items():
    print(f"  {key}: {value}")
````

## 7. Conditionals — `if / elif / else` 🔀

Conditionals let your program make **decisions** based on conditions.

```
if condition:
    # runs if condition is True
elif another_condition:
    # runs if the first was False but this is True
else:
    # runs if all conditions were False
```

````python
score = 78

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
````

**✏️ Try it:** Change `score` to different values (e.g., 95, 55, 83) and observe the output.

## 8. Loops — `for` and `while` 🔁

Loops let you **repeat** actions without rewriting code.

- **`for` loop** — iterate over a sequence (list, string, range)
- **`while` loop** — repeat as long as a condition is true

````python
# for loop over a list
medications = ["Aspirin", "Metformin", "Lisinopril"]

print("Medications prescribed:")
for med in medications:
    print(f"  - {med}")
````

````python
# for loop with range()
print("Countdown:")
for i in range(5, 0, -1):   # start=5, stop=0, step=-1
    print(f"  {i}...")
print("  🚀 Launch!")
````

````python
# while loop
balance = 100
day = 1

print("Daily balance tracker:")
while balance > 0:
    balance -= 30   # Subtract 30 each day
    print(f"  Day {day}: ${max(balance, 0)} remaining")
    day += 1
    if balance <= 0:
        print("  ⚠️  Balance depleted!")
````

## 9. Functions 🔧

A **function** is a reusable block of code that performs a specific task.

```python
def function_name(parameters):
    # code block
    return result
```

````python
# Define a simple function
def greet(name):
    """Return a personalized greeting."""
    return f"Hello, {name}! Welcome to Python."

# Call the function
print(greet("Alice"))
print(greet("Bob"))
````

````python
# Function with multiple parameters and a default value
def calculate_bmi(weight_kg, height_m, name="Patient"):
    """Calculate BMI and return a classification."""
    bmi = weight_kg / (height_m ** 2)
    
    if bmi < 18.5:
        category = "Underweight"
    elif bmi < 25:
        category = "Normal weight"
    elif bmi < 30:
        category = "Overweight"
    else:
        category = "Obese"
    
    return bmi, category

# Call the function
bmi_value, bmi_category = calculate_bmi(70, 1.75, "Sara")
print(f"BMI: {bmi_value:.1f} → {bmi_category}")

bmi_value, bmi_category = calculate_bmi(95, 1.70)
print(f"BMI: {bmi_value:.1f} → {bmi_category}")
````

> 🔍 **Note:** The text inside `"""triple quotes"""` right after `def` is called a **docstring** — it documents what the function does. Good practice!

## 10. 🏁 Mini Project — Patient Summary Generator

Let's combine everything we've learned into a small, practical program.

**Goal:** Write a function that takes patient data and prints a formatted summary report.

````python
def generate_patient_summary(patient):
    """
    Generate a formatted summary for a patient dictionary.
    Expected keys: name, age, weight_kg, height_m, diagnoses, medications
    """
    print("=" * 40)
    print("       PATIENT SUMMARY REPORT")
    print("=" * 40)
    
    # Basic info
    print(f"Name : {patient['name']}")
    print(f"Age  : {patient['age']} years")
    
    # BMI calculation
    bmi = patient['weight_kg'] / (patient['height_m'] ** 2)
    print(f"BMI  : {bmi:.1f}")
    
    # Diagnoses
    print("\nDiagnoses:")
    for i, dx in enumerate(patient['diagnoses'], start=1):
        print(f"  {i}. {dx}")
    
    # Medications
    print("\nMedications:")
    for med in patient['medications']:
        print(f"  - {med}")
    
    # Risk flag
    if patient['age'] > 60 or bmi > 30:
        print("\n⚠️  HIGH RISK — Flag for review")
    else:
        print("\n✅  Standard follow-up")
    
    print("=" * 40)


# --- Test the function ---
patient_1 = {
    "name": "Ahmed Al-Rashid",
    "age": 65,
    "weight_kg": 88,
    "height_m": 1.72,
    "diagnoses": ["Type 2 Diabetes", "Hypertension"],
    "medications": ["Metformin 500mg", "Lisinopril 10mg", "Aspirin 81mg"]
}

patient_2 = {
    "name": "Lina Hassan",
    "age": 32,
    "weight_kg": 62,
    "height_m": 1.65,
    "diagnoses": ["Asthma"],
    "medications": ["Salbutamol inhaler"]
}

generate_patient_summary(patient_1)
print()
generate_patient_summary(patient_2)
````
