## 1. Your first Python program
Python is often introduced with a simple message:

````python
print("Hello, world!")
````


### Principle
Python code is meant to be **readable**.  
A simple command like `print()` lets you display information.


## 2. Variables
Variables store information so you can use it later.

````python

name = "Lina"
age = 20
height = 1.68

print("Name:", name)
print("Age:", age)
print("Height:", height)
````


### Principle
Choose variable names that clearly describe what they hold.  
Good names make code easier to understand.


## 3. Basic data types
Python has different kinds of values.

````python

student_name = "Omar"      # string
score = 95                # integer
temperature = 23.5        # float
is_present = True         # boolean

print(type(student_name))
print(type(score))
print(type(temperature))
print(type(is_present))
````


### Common types
- `str` for text
- `int` for whole numbers
- `float` for decimal numbers
- `bool` for `True` or `False`


## 4. Math in Python

````python

a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor division:", a // b)
print("Remainder:", a % b)
print("Power:", a ** b)
````


### Principle
Python can be used as a calculator, but it is more powerful because it lets you save results in variables and reuse them.


## 5. Strings and text manipulation

````python

first_name = "Ada"
last_name = "Lovelace"

full_name = first_name + " " + last_name
print(full_name)

message = f"Welcome, {full_name}!"
print(message)

print("Length of full name:", len(full_name))
print("Uppercase:", full_name.upper())
print("Lowercase:", full_name.lower())
````


### Principle
Strings represent text.  
Python gives you many built-in tools for working with text.


## 6. Lists
Lists store multiple items in order.

````python

fruits = ["apple", "banana", "orange"]

print(fruits)
print("First fruit:", fruits[0])

fruits.append("mango")
print("After adding mango:", fruits)

print("Number of fruits:", len(fruits))
````


### Principle
Use a list when you want to keep a collection of related items.


## 7. Dictionaries
Dictionaries store data as **key-value pairs**.

````python

student = {
    "name": "Sara",
    "age": 19,
    "major": "Computer Science"
}

print(student)
print("Student name:", student["name"])
print("Major:", student["major"])
````


### Principle
Use a dictionary when you want to label pieces of information clearly.


## 8. Conditionals
Conditionals let your program make decisions.

````python

grade = 82

if grade >= 90:
    print("Excellent")
elif grade >= 70:
    print("Good job")
else:
    print("Keep practicing")
````


### Principle
Python uses indentation to show blocks of code.  
Indentation is not just style in Python, it is part of the syntax.


## 9. Loops
Loops repeat actions.

````python

for number in range(5):
    print("Number:", number)
````

````python

count = 0

while count < 3:
    print("Count is", count)
    count += 1
````


### Principle
- Use a `for` loop when you know what you want to iterate over
- Use a `while` loop when you want to repeat until a condition changes


## 10. Functions
Functions help you organize and reuse code.

````python

def greet(name):
    return f"Hello, {name}!"

print(greet("Maha"))
print(greet("Yousef"))
````


### Principle
A function should usually do **one clear job**.  
This makes code easier to test, reuse, and understand.


## 11. Input and output

````python

favorite_color = "blue"
print("My favorite color is", favorite_color)
````


In notebooks, it is often easier to set example variables directly rather than rely on `input()` every time.

## 12. Writing clean Python


Good Python code is usually:

1. **Readable**
2. **Simple**
3. **Consistent**
4. **Well named**
5. **Easy to test**

### Example of clearer code
Instead of:
```python
x = 5
y = 7
z = x + y
print(z)
```

You might write:
```python
first_number = 5
second_number = 7
total = first_number + second_number
print(total)
```

## 14. Mini challenge

````python

numbers = [4, 7, 2, 9, 5]

total = 0
for number in numbers:
    total += number

average = total / len(numbers)

print("Numbers:", numbers)
print("Total:", total)
print("Average:", average)
````


### Why this matters
This small example combines several Python principles:
- variables
- lists
- loops
- arithmetic
- readable structure
